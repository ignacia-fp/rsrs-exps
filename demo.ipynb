This notebook collects a few common RSRS workflows: single-run setup, shell-script generation, and multi-run sweeps by either `(ref_level, depth)` pairs or mesh width `h`.


```python
from rsrs_config import RSRSBenchmarkConfig
print(RSRSBenchmarkConfig.__init__.__doc__)
```

```python
## Generate the test case either to run the test or retrieve results
## recommendations: if operator_type = 5, and if ref_level = 8, pick depth = 4 (this is a problem of around 500k dofs)
## if ref_level = 9, pick depth = 5 (this is a problem of around 2MM dofs)
config = RSRSBenchmarkConfig(operator_type=5, dim_arg_type=3, ref_level=5, depth=2)
```

```python
## Generate the shell script (disable it unless you want to run the test)
config.generate_bash_script("run_test.sh")
```

## Generate multi-run shell scripts

Use `generate_bash_script_multi(...)` when you want one script that sweeps over thread counts together with a dof marker.

- For `dim_arg_type=3`, pass explicit `(ref_level, depth)` pairs.
- For `dim_arg_type=1` or `2`, pass mesh widths and set `dof_marker="h"`.


```python
## Example: one script that sweeps over explicit (ref_level, depth) pairs and thread counts
config.generate_bash_script_multi(
    filename="run_ref_level_depth_sweep.sh",
    json_num_threads=[4, 8, 16],
    refinement_level_depth_pairs=[(5, 2), (6, 3), (7, 4)],
)
```

```python
## Example: sweep by mesh width h instead of refinement level
meshwidth_config = RSRSBenchmarkConfig(
    operator_type=19,
    dim_arg_type=1,
    precision=1,
    geometry=8,
    h=1.0e-2,
    depth=2,
    kappa=3.0,
    solve=True,
    plot=False,
    save_samples=False,
    load_samples=False,
)

meshwidth_config.generate_bash_script_multi(
    filename="run_meshwidth_sweep.sh",
    json_num_threads=[4, 8, 16],
    mesh_widths=[2.0e-2, 1.0e-2, 5.0e-3],
    dof_marker="h",
)
```
